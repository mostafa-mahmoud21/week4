```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("GlobalLandTemperaturesByCountry.csv")  # Replace with your actual file name
print(df.head())  

# Basic info about the dataset
print(df.info())
print(df.describe(include='all'))

# Plot Histograms 
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Histogram for {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

#  Violin and swarm plots 
cat_cols = df.select_dtypes(include='object').columns
if len(cat_cols) > 0:
    category = cat_cols[0]
    for col in numeric_cols:
        plt.figure(figsize=(10, 4))
        sns.violinplot(x=df[category], y=df[col])
        plt.title(f"Violin plot of {col} by {category}")
        plt.show()

        plt.figure(figsize=(10, 4))
        sns.swarmplot(x=df[category], y=df[col], size=2)
        plt.title(f"Swarm plot of {col} by {category}")
        plt.show()

# Bar plot of categorical columns
for col in cat_cols:
    plt.figure(figsize=(10, 4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count of categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

# Mean, Median, Mode for each numeric column
for col in numeric_cols:
    print(f"\nStats for {col}:")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
    print("Min:", df[col].min())
    print("Max:", df[col].max())

# Check for outliers using boxplots
for col in numeric_cols:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot to check outliers in {col}")
    plt.show()
```

```python
# 1. Are the data what you expect? Are they usable?
# Yes, the data looks usable. The columns are complete, and the values are understandable. Some cleaning might be needed if there are outliers or missing values.

# 2. What is the overall shape of the distribution?
# Most columns show a normal or right-skewed shape. 

# 3. Are there any outliers present?
# Yes, the boxplots show some outliers (points far from the rest). 

# 4. If there are multiple related histograms, how does the distribution change across groups?
# The violin and swarm plots show how values change based on categories. Some groups have higher or wider ranges than others.

# 5. What are the minimum and maximum values in each histogram?
# These were printed using .min() and .max() — each column has its own range of values from the smallest to the largest.

# 6. How do bin sizes affect the histogram?
# Smaller bin sizes show more detail, but can look noisy. Larger bins are smoother but hide smaller patterns. Bin size affects what you can see.

# 7. Does the distribution appear normal, or something else?
# Some columns are normal (bell-shaped), while others are skewed or have multiple peaks (bimodal). It depends on the feature.
```
